On nous informe que les objets proviennent de deux entrepﺃ٤ts diffﺃﺏrents, identifiﺃﺏs par le prﺃﺏfixe de leur `id` :

AM-... : Entrepﺃ٤t d'Amsterdam.

La partie alphanumﺃﺏrique qui suit est le vﺃﺏritable identifiant des objets.

Nous souhaitons nettoyer les identifiants afin de pouvoir trier les objets en fonction de leur vﺃﺏritables identifiants. Il s'agit donc de renvoyer l'information de l'entrepﺃ٤t d'origine ﺃ  une autre colonne.

1. Crﺃﺏez une nouvelle colonne `entrepot`. Remplissez-la avec la valeur "Amsterdam" si l'id commence par "AM-", et "Berlin" s'il commence par "BER-".

Indice :  Vous pouvez utiliser la fonction `np.where` (**[documentation](https://numpy.org/devdocs/reference/generated/numpy.where.html)**).

```python
## ﺃ┬crivez votre code ici
df["entrepot"] = np.where( df["id"].str.startswith("AM-"),"Amsterdam",
    "Berlin"
)
## alternative du prof : df["entrepot"] = np.where(df["id"].str[0]=='A',"Amsterdam", "Berlin")
```

```python
## ﺃ┬crivez votre code ici.
df["id"]=df["id"].str.split('-').str[1]
##voir ce qu 'est split
```
